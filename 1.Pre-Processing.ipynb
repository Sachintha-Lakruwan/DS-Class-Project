```python
def clean_general_ecommerce_platforms(value):
    if pd.isna(value):
        return 'None'

    value = str(value).lower().strip()

    # Handle junk/general responses
    junk_values = [
        'none', 'ikman.lk, ', 'daraz.lk, ', 'https://slbooks.lk/',
        'used some platforms to purchase books and electronic items.',
        'online shopping websites of supermarkets and clothing stores',
        'daraz.lk, online shopping websites of supermarkets and clothing stores',
        'online clothing stores '
    ]
    if value in junk_values or 'used some platforms' in value or 'online shopping' in value:
        return 'None'

    # Fix known typos and normalize names
    replacements = {
        'ali express.com': 'aliexpress',
        'ali express': 'aliexpress',
        'aliexpress.com': 'aliexpress',
        'amazon ': 'amazon',
        'wishique': 'wishque',
        'vishq': 'wishque',
        'wishique ': 'wishque',
        'ebay ': 'ebay',
        'ebay': 'ebay',
        'pickme': 'pickme.lk',
        'instagram stores': 'instagram',
        'slbooks.lk': 'slbooks.lk',
        'shein': 'shein'
    }

    for wrong, correct in replacements.items():
        value = value.replace(wrong, correct)

    # Clean extra commas and spaces
    parts = [part.strip() for part in value.split(',') if part.strip() != '']
    
    # Final list of standardized platforms (title case optional)
    cleaned_parts = []
    for part in parts:
        part_cleaned = part.lower().strip()
        if part_cleaned == 'none' or not part_cleaned:
            continue
        cleaned_parts.append(part_cleaned)

    # Deduplicate and title-case
    cleaned_unique = sorted(set([p.title() for p in cleaned_parts]))

    return ', '.join(cleaned_unique) if cleaned_unique else 'None'

# Apply it
df['General E-Commerce Platforms'] = df['General E-Commerce Platforms'].apply(clean_general_ecommerce_platforms)

# Optional: View results
print(df['General E-Commerce Platforms'].unique())
```

```python
def clean_specialty_platforms(value):
    if pd.isna(value):
        return 'None'

    # Lowercase & strip spaces
    value = str(value).lower().strip()

    # Replace junk values with 'None'
    if value in ['none', 'no', 'n/a', 'no experience in this flatform', 'none ']:
        return 'None'

    # Fix common typo
    value = value.replace('lankavechicle.com', 'lankavehicle.com')

    # Standardize platform name casing
    platforms = {
        'patpat.lk': 'Patpat.lk',
        'riyasewana.lk': 'Riyasewana.lk',
        'lankavehicle.com': 'LankaVehicle.com',
        'autolanka.com': 'AutoLanka.com',
        'pricelanka.lk': 'PriceLanka.lk',
        'ikman.lk': 'Ikman.lk',
        'amazon': 'Amazon',
        'toyota': 'Toyota'
    }

    # Split combined entries
    parts = [part.strip() for part in value.split(',') if part.strip()]
    
    cleaned_parts = []
    for part in parts:
        cleaned = platforms.get(part, part.title())
        if cleaned not in cleaned_parts:
            cleaned_parts.append(cleaned)

    return ', '.join(cleaned_parts) if cleaned_parts else 'None'

# Apply the cleaning function
df['Specialty Online Stores / automobile'] = df['Specialty Online Stores / automobile'].apply(clean_specialty_platforms)

# Optional: check unique values after cleaning
print(df['Specialty Online Stores / automobile'].unique())
```

```python
def clean_online_pharmacies(value):
    if pd.isna(value):
        return 'None'

    value = str(value).lower().strip()

    # Handle junk/irrelevant values
    junk_values = [
        'none', 'no', 'no ', 'n/a', 'nearby', 'pharmacy', 'none ',
        "didn't order any pharmacy items online"
    ]
    if value in junk_values or 'nearby' in value:
        return 'None'

    # Fix typos
    value = value.replace('healthgurd.lk', 'healthguard.lk')
    value = value.replace('union chemist', 'unionchemist.lk')  # Assuming standardization
    value = value.replace('pharmacy ', 'pharmacy')

    # Split entries by comma
    parts = [part.strip() for part in value.split(',') if part.strip() != '']

    # Final platform list standardization
    cleaned_parts = []
    for part in parts:
        if part == 'none' or not part:
            continue
        cleaned_parts.append(part)

    # Deduplicate and title-case nicely
    cleaned_unique = sorted(set([p.title() for p in cleaned_parts]))

    return ', '.join(cleaned_unique) if cleaned_unique else 'None'

# Apply it
df['Online pharmacies'] = df['Online pharmacies'].apply(clean_online_pharmacies)

# Optional: See cleaned unique values
print(df['Online pharmacies'].unique())
```

```python
def clean_fashion_beauty(value):
    if pd.isna(value):
        return 'None'

    value = str(value).lower().strip()

    # Junk responses
    junk_values = ['none', 'n/a', "didn\x92t used", '']
    if value in junk_values or 'didn' in value:
        return 'None'

    # Standardize known names
    replacements = {
        'kellf felder': 'kelly felder',
        'kellyfelder': 'kelly felder',
        'kapruka.com/fashion': 'kapruka.com/fashion',
        'nolimit.lk / gflock.lk': 'nolimit.lk, gflock.lk',
        'zigzag.lk': 'zigzag.lk',
        'tharshana and insta shops': 'tharshana, instagram',
        's&s': 'spring and summer',
        'gflock': 'gflock.lk',
        'mimosa': 'mimosa.lk',
        'chenara dodge': 'chenara dodge',
        'dsi': 'dsi.lk',
        'carlo clothing': 'carlo clothing',
        'moose': 'moose.lk',
        'moods': 'moods.lk',
        'amanthe': 'amanthe.lk',
        'noorareed.lk': 'noorareed.lk'
    }

    for wrong, correct in replacements.items():
        value = value.replace(wrong, correct)

    # Split by commas or slashes
    value = value.replace('/', ',')
    parts = [p.strip() for p in value.split(',') if p.strip()]

    cleaned_parts = sorted(set([p.title() for p in parts]))

    return ', '.join(cleaned_parts) if cleaned_parts else 'None'

# Apply it
df['Fashion and beauty retailers'] = df['Fashion and beauty retailers'].apply(clean_fashion_beauty)

# Optional: View cleaned results
print(df['Fashion and beauty retailers'].unique())
```

```python
def clean_grocery_services(value):
    if pd.isna(value):
        return 'None'

    value = str(value).lower().strip()

    # Handle junk responses
    junk_values = ['none', 'n/a', '']
    if value in junk_values:
        return 'None'

    # Fix known typos or variations
    replacements = {
        'keels ': 'keellssuper.lk',
        'glomark': 'glomark.lk',
        'glomark ': 'glomark.lk',
        'glomark.': 'glomark.lk',
        'cargillis food city': 'cargillsonline.com',
        'cargillis': 'cargillsonline.com',
        'cargillsonline.com': 'cargillsonline.com',
        'keels': 'keellssuper.lk',
        'cinnamonhotel.com': 'cinnamonhotels.com'
    }

    for wrong, correct in replacements.items():
        value = value.replace(wrong, correct)

    # Split and clean
    parts = [p.strip() for p in value.split(',') if p.strip()]
    cleaned_parts = sorted(set([p.title() for p in parts]))

    return ', '.join(cleaned_parts) if cleaned_parts else 'None'

# Apply it
df['Grocery delivery services'] = df['Grocery delivery services'].apply(clean_grocery_services)

# Optional: check result
print(df['Grocery delivery services'].unique())
```
